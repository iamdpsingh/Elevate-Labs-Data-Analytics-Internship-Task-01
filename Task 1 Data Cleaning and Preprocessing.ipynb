```python
import pandas as pd
```

```python
#input data
df = pd.read_csv("Customer Personality Analysis dataset\\marketing_campaign.csv", sep="\t")
```

```python
print("Initial Shape:", df.shape)
```

```python
df.info()
```

```python
df.head()
```

```python
df.columns=df.columns.str.strip()
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
df=df.rename(columns={"id": "Customer_id"})
df
```

```python
print("\nMissing values:\n",df.isnull().sum())
df=df.dropna(subset=["income"])  # rows with missing values will be dropped
```

```python
df=df.drop_duplicates() # will remove duplicate values
```

```python
df["education"]=df["education"].str.strip().str.title()
df["marital_status"]=df["marital_status"].str.strip().str.title()

# This is for if we want to standardize the text one by one
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip().str.title()

# This is for if we want to standardize the text through loop in which all object(string) columns and clean them
```

```python
# This will set the date format

df["dt_customer"]=pd.to_datetime(df["dt_customer"],format='%d-%m-%Y')
df["dt_customer"]=df["dt_customer"].dt.strftime("%d-%m-%Y")
```
